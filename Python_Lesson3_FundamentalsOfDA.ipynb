## Basic data cleaning / pre-processing

### 7. Identify any blank or erroneous values

```python
#Identify blank values

nan_df = df.isna() # using isna() on the 'df' will return the same dataframe shape as 'df', but, will replace
                   # each value with a 'True' / 'False' value; returning 'True' when a cell is blank, else 'False'.


# Identify Erroneous values in Sepal.Length, and coerce those values to NaN
df.loc[:,'Sepal.Length'] = pd.to_numeric(df['Sepal.Length'],errors='coerce')
```

### 8. Identify duplicate values

```python
# Identify duplicates
df['Duplicates'] = df.duplicated() # Create a column which returns 'True' or 'False' depending on if there are duplicates

df.loc[df['Duplicates']==True] # Display duplicate values in 'df'
```

### 9. Remove or transform blank, duplicate, or erroneous rows

```python
df = df.dropna() # We can drop rows containing blanks by using 'dropna()' on the df.
df.drop_duplicates(inplace=True) # drop duplicates like this
df # display the df with data 'cleaned'
```

### 10. Drop unnecessary columns

<div class="alert alert-block alert-info">
<b>Tip:</b> Removing unnecessary data from your dataset can significantly reduce processing time, particularly on larger datasets.
</div>

```python
df.drop(columns = ['RowNum', 'Species2', 'Duplicates'], inplace=True) # Drop the columns which aren't relevant
df.head(5) # display first 5 rows to show change in df columns

df = df.reset_index(drop=True)
```

### 11. Create a Seaborn 'joint plot' presenting the length by width of the Iris flower's Sepal

```python
ax=sns.jointplot(x='Sepal.Length',y='Sepal.Width',data=df,height=5)
```

### 12. Create a Seaborn 'strip plot' presenting the Sepal length of the Iris flower, by species

```python
ax=sns.stripplot(x='Species',y='Sepal.Length',data=df,jitter=True,edgecolor='gray')
```

### 13. Plot a Seaborn 'pair plot' showing the correlation between the length and width of Iris flowers


<div class="alert alert-block alert-info">
<b>Tip:</b> Notice that there is significant skewing in a particular column suggesting a potentially erroneous outlier. This visual representation is another good way to identify data quality issues.
</div>

```python
ax=sns.pairplot(data=df,kind='scatter', hue='Species')
```
